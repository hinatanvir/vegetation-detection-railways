```python
#Creating a dataset for EDA
# First creates lists of labels 
#path_images = '/Users/arasdirekoglu/Documents/Studium/Master/HTW - MPMD/3. Semester/MPMD 3.1. PM & Data Analytics Lab/02_Project/00_data/rs19_val/uint8/rs19_val/'

list_files = []
list_road = [] # 0
list_sidewalk=[] # 1 
list_construction = [] # 2 
list_tram_track = [] # 3 
list_fence = [] # 4
list_pole = [] # 5 
list_traffic_light = [] # 6
list_traffic_sign = [] # 7
list_vegetation = [] # 8 
list_terrain = [] # 9
list_sky = [] # 10
list_human = [] # 11
list_rail_track = [] # 12 
list_car = [] # 13
list_truck = [] # 14
list_trackbed = [] #15
list_onrails = [] #16 
list_rail_raised = [] #17
list_rail_embedded = [] #18


for filename in os.listdir(path_images):
    if filename.endswith('.png'):
        mask = cv2.imread(path_images+filename)
        label = np.unique(mask[:,:,0])
        list_files.append(filename) 
        for i in label:
            if i == 0:
                list_road.append(filename)
            elif i == 1:
                list_sidewalk.append(filename)
            elif i == 2:
                list_construction.append(filename)
            elif i == 3: 
                list_tram_track.append(filename)
            elif i == 4:
                list_fence.append(filename)
            elif i == 5:
                list_pole.append(filename)
            elif i == 6:
                list_traffic_light.append(filename)
            elif i == 7:
                list_traffic_sign.append(filename)
            elif i == 8:
                list_vegetation.append(filename)
            elif i == 9:
                list_terrain.append(filename)
            elif i == 10:
                list_sky.append(filename)
            elif i == 11:
                list_human.append(filename)
            elif i == 12:
                list_rail_track.append(filename)
            elif i == 13:
                list_car.append(filename)
            elif i == 14:
                list_truck.append(filename)
            elif i == 15:
                list_trackbed.append(filename)
            elif i == 16:
                list_onrails.append(filename)
            elif i == 17:
                list_rail_raised.append(filename)
            elif i == 18:
                list_rail_embedded.append(filename)
```

```python
#Creating a Dataframe out of filename
df = pd.DataFrame(list_files, columns=['filename'])

#Add 1 if image has label, add 0 if it does not
#Road
roadtrue = []
for i in df.filename:
    if i in list_road:
        roadtrue.append(1)
    else:
        roadtrue.append(0) 
df['road'] = roadtrue

#sidewalk
sidewalktrue = []
for i in df.filename:
    if i in list_sidewalk:
        sidewalktrue.append(1)
    else:
        sidewalktrue.append(0) 
df['side_walk'] = sidewalktrue

#construction
constructiontrue = []
for i in df.filename:
    if i in list_construction:
        constructiontrue.append(1)
    else:
        constructiontrue.append(0) 
df['construction'] = constructiontrue

#tramtrack
tramtracktrue = []
for i in df.filename:
    if i in list_tram_track:
        tramtracktrue.append(1)
    else:
        tramtracktrue.append(0) 
df['tram_track'] = tramtracktrue

#fence
fencetrue = []
for i in df.filename:
    if i in list_fence:
        fencetrue.append(1)
    else:
        fencetrue.append(0) 
df['fence'] = fencetrue

#pole
poletrue = []
for i in df.filename:
    if i in list_pole:
        poletrue.append(1)
    else:
        poletrue.append(0) 
df['pole'] = poletrue

#traffic light
trafficlighttrue = []
for i in df.filename:
    if i in list_traffic_light:
        trafficlighttrue.append(1)
    else:
        trafficlighttrue.append(0) 
df['traffic_light'] = trafficlighttrue

#traffic sign
trafficsigntrue = []
for i in df.filename:
    if i in list_traffic_sign:
        trafficsigntrue.append(1)
    else:
        trafficsigntrue.append(0) 
df['traffic_sign'] = trafficsigntrue

#vegetation
vegetationtrue = []
for i in df.filename:
    if i in list_vegetation:
        vegetationtrue.append(1)
    else:
        vegetationtrue.append(0) 
df['vegetation'] = vegetationtrue

#terrain
terraintrue = []
for i in df.filename:
    if i in list_terrain:
        terraintrue.append(1)
    else:
        terraintrue.append(0) 
df['terrain'] = terraintrue

#sky
skytrue = []
for i in df.filename:
    if i in list_sky:
        skytrue.append(1)
    else:
        skytrue.append(0) 
df['sky'] = skytrue

#human
humantrue = []
for i in df.filename:
    if i in list_human:
        humantrue.append(1)
    else:
        humantrue.append(0) 
df['human'] = humantrue

#rail_track
rail_tracktrue = []
for i in df.filename:
    if i in list_rail_track:
        rail_tracktrue.append(1)
    else:
        rail_tracktrue.append(0) 
df['rail_track'] = rail_tracktrue

#car
cartrue = []
for i in df.filename:
    if i in list_car:
        cartrue.append(1)
    else:
        cartrue.append(0) 
df['car'] = cartrue

#truck
trucktrue = []
for i in df.filename:
    if i in list_truck:
        trucktrue.append(1)
    else:
        trucktrue.append(0) 
df['truck'] = trucktrue

#trackbed
trackbedtrue = []
for i in df.filename:
    if i in list_trackbed:
        trackbedtrue.append(1)
    else:
        trackbedtrue.append(0) 
df['trackbed'] = trackbedtrue

#onrails
onrailstrue = []
for i in df.filename:
    if i in list_onrails:
        onrailstrue.append(1)
    else:
        onrailstrue.append(0) 
df['on_rails'] = onrailstrue

#rail_raised
railraisedtrue = []
for i in df.filename:
    if i in list_rail_raised:
        railraisedtrue.append(1)
    else:
        railraisedtrue.append(0) 
df['rail_raised'] = railraisedtrue

#rail_embedded
railembeddedtrue = []
for i in df.filename:
    if i in list_rail_embedded:
        railembeddedtrue.append(1)
    else:
        railembeddedtrue.append(0) 
df['rail_embedded'] = railembeddedtrue
```
